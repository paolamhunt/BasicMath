```python
# Print the names of the columns in this DataFrame.
print(batting.columns.values)
```

```python
# Create a DataFrame named total_hr
# Select the teamID, yearID, and HR columns from batting
# Group the results by both teamID and yearID. set as_index perameter to False
# Summing the remaining column within each group
# Sort the values by the total number of home runs, in descending order

total_hr = batting.loc[:,['teamID','yearID', 'HR']
                    ].groupby(['teamID', 'yearID'], as_index=False
                             ).sum().sort_values(by='HR', ascending=False
                                                  )
# display the first 10 rows of this DataFrame
total_hr.head(10)
```

```python
# Create a DataFrame named avg_hr
# Select the teamID and HR columns from the total_hr DataFrame
# Group the results by teamID
# Average the remaining column within each group
# Sort the values by the average number of home runs per season, in descending order

avg_hr = total_hr.loc[:,['teamID', 'HR']
                    ].groupby('teamID'
                             ).mean().sort_values(by='HR', ascending=False
                                                  )
# Use head() to display the first 10 rows of this DataFrame
avg_hr.head(10)
```

## Problem 7: Batting Data – Batting Averages


Then add a new column named BA to the batting_avg DataFrame. This new column should be calculated by dividing the values in the H column by the values in the AB column. This should be done without using a loop.
Finally, sort the DataFrame by BA, in descending order. Use head() to display the first 10 rows of this DataFrame.

```python
# Create a DataFrame named batting_avg
# Select teamID, yearID, H, and AB columns from the batting DataFrame
# Group results by teamID and yearID
# Calculate the sum of the remaining two columns within each group. 

batting_avg = batting.loc[:,['teamID','yearID', 'H','AB']
                    ].groupby(['teamID', 'yearID']
                             ).sum()

# Add a new column named BA to the batting_avg DataFrame
# Without using a loop, calculate new col by dividing the values in the H column by the values in the AB column
# Sort the DataFrame by BA, in descending order
batting_avg.loc[:,'BA'] = batting_avg.loc[:,'H'] / batting_avg.loc[:,'AB']
batting_avg=batting_avg.sort_values(by='BA', ascending=False)

# Display the first 10 rows of this DataFrame
batting_avg.head(10)
```

## Problem 8: Batting Data – Cardinals vs. Cubs

```python
# Create a DataFrame named stl_batting
# Use loc to filter batting DF; keeping only SNL team and year >= 1900
# Use loc to select the yearID, H, AB, and HR
# Group the results by yearID
# Calculate grouped sums

stl_sel = (batting.loc[:,'teamID'] == 'SLN') & (batting.loc[:,'yearID'] >= 1900)
stl_batting = batting.loc[stl_sel,['yearID','H','AB','HR']
                         ].groupby('yearID'
                                  ).sum()

# Add a new column named BA to the stl_batting DataFrame
# Without using a loop, calculate new col by dividing the values in the H column by the values in the AB column
# Sort the DataFrame by BA, in descending order
stl_batting.loc[:,'BA'] = stl_batting.loc[:,'H'] / stl_batting.loc[:,'AB']


# Create a DataFrame named chi_batting
# Use loc to filter batting DF; keeping only CHN team and year >= 1900
# Use loc to select the yearID, H, AB, and HR
# Group the results by yearID
# Calculate grouped sums

chi_sel = (batting.loc[:,'teamID'] == 'CHN') & (batting.loc[:,'yearID'] >= 1900)
chi_batting = batting.loc[chi_sel,['yearID','H','AB','HR']
                         ].groupby('yearID'
                                  ).sum()

# Add a new column named BA to the chi_batting DataFrame
# Without using a loop, calculate new col by dividing the values in the H column by the values in the AB column
# Sort the DataFrame by BA, in descending order
chi_batting.loc[:,'BA'] = chi_batting.loc[:,'H'] / chi_batting.loc[:,'AB']

```

```python
plt.figure(figsize=[12,4])

plt.subplot(1,2,1)
plt.plot(stl_batting.index, stl_batting.loc[:, 'BA'], color='red', label='Cardinals')
plt.plot(chi_batting.index, chi_batting.loc[:, 'BA'], color='blue', label='Cubs')

plt.legend()
plt.xlabel("Year")
plt.ylabel("Batting Average")
plt.title('Batting Average By Year')

plt.subplot(1,2,2)
plt.plot(stl_batting.index, stl_batting.loc[:,'HR'], color='red', label='Cardinals')
plt.plot(chi_batting.index, chi_batting.loc[:,'HR'], color='blue', label='Cubs')

plt.legend()
plt.xlabel("Year")
plt.ylabel("Home Runs")
plt.title('Home Runs by Year')

plt.show()
```

```python
BAresult = round(np.mean(stl_batting.loc[:,'BA'] > chi_batting.loc[:,'BA']),4)
print(BAresult)
```
